```python
# Loading libraries
import datetime as dt
import numpy as np
import pandas as pd
import yfinance as yf
from datetime import timedelta
from statistics import mean

# Creating a list of assets:
my_tickers_1 = ["META", "CROX", "AMZN", "OXY", "NFLX", "DECK", "URI"]
my_tickers_1 = sorted(my_tickers_1, reverse=False)  # Sorting alphabetically

my_tickers_2 = ["SPY", "IWM", "DIA"]  # ETFs
my_tickers_2 = sorted(my_tickers_2, reverse=False)  # Sorting alphabetically

my_tickers_3 = my_tickers_1+my_tickers_2 # Used for Beta calculation later
my_tickers_3 = sorted(my_tickers_3, reverse=False)  # Sorting alphabetically

start_date = dt.datetime.today() - dt.timedelta(days=365)
end_date = dt.datetime.today()
```

```python
## Now downloading the data using the above variables
my_port = yf.download(start=start_date, end=end_date, tickers=my_tickers_1)
etf = yf.download(start=start_date, end=end_date, tickers=my_tickers_2)
combined_port = yf.download(start=start_date, end=end_date, tickers=my_tickers_3)
my_port # Visualizing stock data
```

```python
# Only tracking the Adj Close values for each stock and dropping NAN values.
my_port = my_port["Adj Close"].dropna()
etf = etf["Adj Close"].dropna()
combined_port = combined_port["Adj Close"].dropna()
```

# Part 1
Using Pandas Dataframe to create a table of stock information.

```python
#  Adding the Tickers Column
data = {"Tickers": my_tickers_1}
table = pd.DataFrame(data)
print(table)
```

```python
# Adding the Portfolio Weight Column
weight = 1/len(my_tickers_1) # Distributed equally between each stock
table["Portfolio Weight"] = weight
print(table)
```

```python
# Calculating Annualized Voltaility using trailing 3 months
start_date = dt.date.today() - dt.timedelta(days=90)
end_date = dt.date.today()

# Annualized Volatility's formula was taken from this website: https://www.fool.com/knowledge-center/how-to-calculate-annualized-volatility.aspx
new_port = (my_port.loc[start_date:end_date].pct_change().std()) * np.sqrt(252) # Taking standard deviation of daily datapoints(percent change) and multiplying by sqrt(252) for annualized data. 
# To filter the relevant dates, I used this stack exchange as reference: https://stackoverflow.com/questions/22898824/filtering-pandas-dataframes-on-dates
listObj = new_port.tolist() # New port is a Series datatype, so using this line to convert to a list.
table["Annual Volatility"] = listObj
print(table)
```

```python
# Calculating Beta against ETFs
covariance = combined_port.pct_change().cov() # Calculating the covariance matrix
variance_spy = combined_port["SPY"].pct_change().var() # Finding variance of eacb ETFs
variance_iwm = combined_port["IWM"].pct_change().var()
variance_dia = combined_port["DIA"].pct_change().var()
spy_index = combined_port.columns.get_loc('SPY') # Finding the column index of ETFs so I can locate the matrix element.
iwm_index = combined_port.columns.get_loc('IWM')
dia_index = combined_port.columns.get_loc('DIA')
beta_spy = list() # Creates empty list
beta_iwm = list()
beta_dia= list()
for x in range(len(my_tickers_3)): # Loop over tickers
    if my_tickers_3[x] != 'SPY' and my_tickers_3[x] != 'IWM' and my_tickers_3[x] != 'DIA': # Using this to avoid diagonal terms
        beta1= covariance.iloc[x,spy_index]/variance_spy # using this to access off-diagonal terms which, when divided by the variance of the ETF, gives me beta ratio against the ETFs
        beta_spy.append(beta1)
        beta2= covariance.iloc[x,iwm_index]/variance_iwm
        beta_iwm.append(beta2)
        beta3= covariance.iloc[x,dia_index]/variance_dia
        beta_dia.append(beta3)
table["Beta Against SPY"] = beta_spy
table["Beta Against IWM"] = beta_iwm
table["Beta Against DIA"] = beta_dia
table
```

```python
# Calculating Average Weekly Drawdowns

weekly_averages_list = list() # Creates empty list
start_date = dt.date.today()
weekly_drawdowns_list = list() # Creates empty list
for y in range(len(my_tickers_1)): # Loop over tickers
    for z in range(1,52): # Loop over 52 weeks
        max_value = my_port[my_tickers_1[y]].loc[start_date-dt.timedelta(days=6):start_date].max() # Gives max value in a one week interval
        min_value = my_port[my_tickers_1[y]].loc[start_date-dt.timedelta(days=6):start_date].min() # Gives min value in a one week interval
        weekly_drawdowns_list.append((min_value-max_value)/max_value) # Appending weekly drawdown to a list 
        start_date = start_date - dt.timedelta(days=6) # Going to the next 7 day interval
    weekly_averages_list.append(mean(weekly_drawdowns_list)) # Taking averages and adding to a list. Individual values correspond to average weekly drawdown of each stock.
    weekly_drawdowns_list = list() # Initializing
    start_date = dt.date.today() # Initializing to repeat the same procedure with another stock
table["Average Weekly Drawdown "] = weekly_averages_list
table
```

```python
# Calculating Maximum Weekly Drawdown
start_date = dt.date.today() 
weekly_drawdowns_list=list()
max_weekly_drawdrowns_list=list()
for k in range(len(my_tickers_1)):
    for l in range(1,52):
        max_value = my_port[my_tickers_1[k]].loc[start_date-dt.timedelta(days=6):start_date].max() # Gives max value in a one week interval
        min_value = my_port[my_tickers_1[k]].loc[start_date-dt.timedelta(days=6):start_date].min() # Gives min value in a one week interval
        weekly_drawdowns_list.append((min_value-max_value)/max_value)# Appending weekly drawdown to a list 
        start_date = start_date - dt.timedelta(days=6)# Going to the next 7 day interval
    max_weekly_drawdrowns_list.append(min(weekly_drawdowns_list)) # Using min() to find the max drawdowns in the given list. Using min because we're looking for the highest negative magnitude.
    weekly_drawdowns_list = list()# Initializing
    start_date = dt.date.today()# Initializing to repeat the same procedure with another stock
table["Maximum Weekly Drawdown "] = max_weekly_drawdrowns_list
table
```

```python
# Calculating Rate of Return
start_date = dt.datetime.today() - dt.timedelta(days=365*10+2) # 10 years + 2 days accounting for leap year.
end_date = dt.datetime.today()
# Importing data for last 10 years
my_port_10y = yf.download(start=start_date, end=end_date, tickers=my_tickers_1)
my_port_10y = my_port_10y["Adj Close"]
etf_10y = yf.download(start=start_date, end=end_date, tickers=my_tickers_2)
etf_10y = etf_10y["Adj Close"]
```

```python
my_port_10y # Checking
```

```python
# Code to calculate the beginning price of each stocks 10 years ago. Copied from the following stack exchange: https://stackoverflow.com/questions/50586146/find-first-non-zero-value-in-each-column-of-pandas-dataframe
beginning_price_list = list()
m = my_port_10y.ne(0).idxmax()
val=pd.DataFrame(dict(pos=m, val=my_port_10y.lookup(m, m.index)))
beginning_price_list = val.iloc[:,1]
beginning_price_list = beginning_price_list.tolist()
val # Checking
```

```python
# Code to calculate the current price of each stocks.
current_price_list = list()
for j in range(len(my_tickers_1)):
        last_data=my_port_10y.iloc[my_port_10y.shape[0]-1,j]
        current_price_list.append(last_data)
        
current_price_list # Checking
        
```

```python
# Calculating total return
Total_return_list = list()
for m in range (len(my_tickers_1)):
    Total_return = ((current_price_list[m]-beginning_price_list[m])/beginning_price_list[m])*100 # Formula to calculate Total Return = (Ending price - Beginning price)/Beginning price
    Total_return_list.append(Total_return)
table["Total Return"] = Total_return_list
table
```

```python
# Calculating Annualized Total Return
annualized_rate_list = list()
for m in range (len(my_tickers_1)):
    annualized_rate = (((current_price_list[m]/beginning_price_list[m])**(1/10))-1)*100 # Using the formula: ((Ending price/Beginning price)^(1/N) -1)*100
    annualized_rate_list.append(annualized_rate)
table["Annualized Total Return"] = annualized_rate_list
```

**Part 2**

```python
# Putting ETF tickers on a table
data = {"Tickers": my_tickers_2}
table2 = pd.DataFrame(data)
print(table2)
```

```python
# Calculating the correlations & covariations against the ETFs using data from 1 year
covariance_list = list()
correlation_list = list()
weighted_return = (my_port.sum(axis = 'columns'))*weight # Summing over rows with appropriate weight calculated in Part 1 to get weighted return.
weighted_return = weighted_return.tolist() # Converting dataframe to list
for x in range(len(my_tickers_2)):
    covariance=(np.cov(etf[my_tickers_2[x]].tolist(),weighted_return))[0,1] # Calculating covariance matrix using np.cov() and accessing off-diagonal term for different ETFs to get covariance of portfolio against ETFs.
    covariance_list.append(covariance)
    correlation=(np.corrcoef(etf[my_tickers_2[x]].tolist(),weighted_return))[0,1]# Calculating correlation matrix using np.corrcoef() and accessing off-diagonal term for different ETFs to get correlation of portfolio against ETFs.
    correlation_list.append(correlation)
table2["Correlation against ETF"]=correlation_list
table2["Covariance against ETF"]=covariance_list
table2
```

```python
# Calculating Tracking Errors using trailing 10-years
weighted_return_10y = (my_port_10y.sum(axis = 'columns'))*weight # Weighted 10 year return
weighted_return_10y = weighted_return_10y.tolist()
standard_deviation_list = list()
for z in range(len(my_tickers_2)):
    standard_deviation=np.std(np.subtract(weighted_return_10y,(etf_10y[my_tickers_2[z]].tolist()))) # Taking standard deviation of the difference between weighted return and ETFs
    standard_deviation_list.append(standard_deviation)
table2["Tracking Errors"]=standard_deviation_list
table2
```

```python
# Calculating Sharpe Ratio using the dataset for 10 years.
weighted_return_annual= ((my_port_10y.sum(axis = 'columns'))*weight).pct_change().mean()*252 # Mean of percent change in weighted return is multiplied by 252 to get annual rate.
risk_free_rate = 0.043; # 4.3 % from a quick google search.
sharpe_ratio_list = list()
for d in range(len(my_tickers_2)):
    sharpe_ratio=(weighted_return_annual-risk_free_rate)/(standard_deviation_list[d]*np.sqrt(252)) # Sharpe Ratio = (Return rate - Risk free rate)/ Standard deviation of excess returns. Excess returns was calculated in the previous part as standard_deviation_list.
    sharpe_ratio_list.append(sharpe_ratio)
table2["Sharpe Ratio"]=sharpe_ratio_list
table2
```

```python
# Annualized Volatility Spread
start_date = dt.date.today() - dt.timedelta(days=252)
end_date = dt.date.today()
annualized_volatility_portfolio = ((my_port.loc[start_date:end_date].sum(axis = 'columns')*weight).pct_change().std()) * np.sqrt(252) # Annualized Volatility of weighted returns
annualized_volatility_spread_list = list()
for g in range(len(my_tickers_2)):
    annualized_volatility_spread = annualized_volatility_portfolio-((etf[my_tickers_2[g]].loc[start_date:end_date].pct_change().std())* np.sqrt(252)) # Annualized Volatility of weighted returns - Annualized Volatility of ETFs
    annualized_volatility_spread_list.append(annualized_volatility_spread)
table2["Annualized Volatility Spread"] = annualized_volatility_spread_list
```
